```python
# Distribution of Age

plt.figure(figsize=(8, 6))
plt.hist(df['AGE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Duration of Stay
plt.figure(figsize=(8, 6))
plt.hist(df['DURATION OF STAY'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of Hospital Stay Duration')
plt.xlabel('Duration of Stay (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Gender Distribution
gender_counts = df['GENDER'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution')
plt.show()
```

```python
#: Time-Series Analysis
df['D.O.A'] = pd.to_datetime(df['D.O.A'], errors='coerce')  # Handle invalid dates
df['Year'] = df['D.O.A'].dt.year
```

```python
# Admissions Over the Years
yearly_admissions = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(yearly_admissions.index, yearly_admissions.values, marker='o', linestyle='--', color='green')
plt.title('Number of Admissions Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.show()
```

```python
# Bivariate Analysis
#Age vs. Duration of Stay
plt.figure(figsize=(8, 6))
plt.scatter(df['AGE'], df['DURATION OF STAY'], alpha=0.5, color='purple')
plt.title('Age vs. Duration of Stay')
plt.xlabel('Age')
plt.ylabel('Duration of Stay (Days)')
plt.show()
```

```python
#Admissions by Gender
gender_yearly = df.groupby(['Year', 'GENDER']).size().unstack()
gender_yearly.plot(kind='bar', figsize=(10, 6), stacked=True, color=['lightblue', 'lightpink'])
plt.title('Yearly Admissions by Gender')
plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.legend(title='Gender')
plt.show()
```

```python
#Handle Missing Data
df_cleaned = df.dropna(subset=['D.O.A'])
```
