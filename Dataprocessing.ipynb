```python
import pandas as pd

```

```python
data={
    'Id':[1,2,3,4,5,6],
    'Age':[23,23,34,24,24,None],
    'Salary':[15000,16000,23000,24000,40000,21000],
    'Department':['HR','HR','IT','Civil','Mechnical','Electrical'],
    'Experience':[2,3,10,2,6,None],
    'Performance':['good','very good','excellent','poor','satisfactory','poor']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
data=df.dropna()
data
```

```python
df=df.fillna(0)
df
```
