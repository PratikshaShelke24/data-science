```python
import pandas as pd
```

```python
df=pd.DataFrame({
        "Event":["trainer1","trainer2","trainer3","trainer4","trainer5"],
        "No_of_Classes":[20,7,26,11,16]

    }
)
```

```python
df
```

```python
#find probability
totalsum=df["No_of_Classes"].sum()
print(totalsum)
```

```python
#find probability
df["probability"]=df.No_of_Classes/totalsum
```

```python
df
```

```python
#sum of probability
df["probability"].sum()
```

```python
#create visulization of probability
import matplotlib.pyplot as plt
```

```python
plt.xlabel("trainer taken by classses")
plt.ylabel("Number of classes")
plt.title("trainer taken by classes")
plt.bar(df.Event,df.probability,width=0.4)
plt.grid(axis='y')
plt.show()
```
