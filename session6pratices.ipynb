```python
import numpy as np
import pandas as pd
from scipy.stats import skew,kurtosis
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 20]
print(type(data))
```

```python
data_skewness = skew(data)
print(f"Skewness: {data_skewness}")

```

```python
plt.hist(data, bins=8, color='skyblue', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#kurtosis
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 20]
```

```python
data_kurtosis = kurtosis(data, fisher=False)
print(f"Kurtosis: {data_kurtosis}")


```

```python
plt.hist(data, bins=8, color='lightgreen', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
