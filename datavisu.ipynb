```python
#matplotlib
import matplotlib.pyplot as plt
```

```python
#sample dataset
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.plot(x,y)
plt.title("sample plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#seaborn
import seaborn as sns
import matplotlib.pyplot as plt

#import scatter dataset
tips=sns.load_dataset("tips")
print(tips.head())

```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=sns.load_dataset("iris"))
plt.show()
```

```python
#plotly
import plotly.express as px
fig=px.scatter(x=[1,2,3,4],y=[10,11,12,13],title="interactive scatter plot")
fig.show()
```

```python
#bar chart
import matplotlib.pyplot as plt
categories=[1,2,3,4,5]
values=[10,21,32,43,53]
plt.bar(categories,values)
plt.title("bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#line graph
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","june","july","aug","sep","oct","nov","dec"]
sales=[20,30,10,40,50,20,30,40,50,70,80,80]
plt.plot(months,sales)
plt.title("line graph")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

```python
#scattter graph
import matplotlib.pyplot as plt
age=[20,24,27,30,35]
income=[15000,20000,25000,27000,35000]
plt.scatter(age,income)
plt.title("scatter graph")
plt.xlabel("age")
plt.ylabel("income")
plt.show()
```

```python
#histogram grap
import matplotlib.pyplot as plt
data=[12,20,30,40,50,60,70,80,90]
plt.hist(data,bins=5)
plt.title("histogram graph")
plt.xlabel("data")
plt.ylabel("frequency")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
labels=['a','b','c','d','e']
size=[10,20,30,40,50]
plt.pie(size,labels=labels,autopct='%1.1f%%')
plt.title("pie chart")
plt.show()
```

```python
#hitmap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 12)  # Random 10x12 matrix
sns.heatmap(data, cmap='coolwarm', annot=True)
plt.title('Heatmap Example')
plt.show()
```
